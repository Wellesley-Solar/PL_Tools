```python
#Imagine a simpler example
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/Karen/Desktop/GitHub/PL_Tools/11-20-2023/example csv for com large.csv'

df = pd.read_csv(file_path)

#Define file length and pixel (group) size
file_length = len(df)
pixel_size = 2 
print('File length is',file_length,'rows.')

#Initiate variables 
frame_number = 1
start = 0 #start at row index 0
end = 2 #end at row index 2 (not included) 


time_stamps = []
com_values = []


#Create conditional loop to go through each group and calculate CoM
while start < file_length:
    int_wvln = 0
    total_int = 0
    for row in range(start,end): 
        wavelength_value = df['Wavelength'][row] 
        intensity_value = df['Intensity'][row]
        int_wvln += wavelength_value * intensity_value
        total_int += intensity_value
        #print('Wvln: ', wavelength_value,'Int: ', intensity_value)
    print('Int x Wvln: ', int_wvln, 'Total intensity: ', total_int)
    center_of_mass = int_wvln / total_int
    print(center_of_mass)
    print('End of frame', frame_number)
    
    #Add the data point (time, com) for the group to the list of x,y values to plot
    time_stamps.append(frame_number)
    com_values.append(center_of_mass)
    #print(time_stamps)
    #print(com_values)
    
    #Update variables 
    frame_number +=1 
    start = (frame_number - 1) * pixel_size
    end = start + pixel_size

print('Time values: ', time_stamps)
print('COM values: ', com_values)
    
plt.plot(time_stamps, com_values, marker='o', linestyle='-',)
plt.xlabel('Time (s)')
plt.ylabel('Center of Mass (nm)')
plt.title('Example File')


```
