This code is for plotting a center of mass v.s. time graph using data from one large csv file (data format for PL from the Belisle Solar Lab). 

The data is stored in one large csv file (an example is shown below). Each data frame is 1340 pixels (rows) long, each row representing a wavelengths from around 400nm - 980 nm. We took 300 frames, with 1 measurement per frame, and 1000ms integration time (meaning that each frame takes 1s to capture). 

### Imagine a simpler example
This chunk of code is meant to help visualize the function of the chunk of code which deals with the actual large file, using a smaller example. It uses a example file which is also in the GitHub folder.

```python
#Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/Karen/Desktop/GitHub/PL_Tools/11-20-2023/example csv for com large.csv'

#Read the csv file into a data frame
df = pd.read_csv(file_path) 

#Define file length and pixel (group) size
file_length = len(df)
pixel_size = 2 
print('File length is',file_length,'rows.') 

#Initiate variables 
frame_number = 1
start = 0 #start at row index 0
end = 2 #end at row index 2 (not included) 


#Create lists which will hold data points (x and y values)
time_stamps = []
com_values = []


#Create conditional loop to go through each group and calculate CoM
while start < file_length:
    #Initiate variables for calculating CoM
    int_wvln = 0
    total_int = 0
    for row in range(start,end): #Loop through each row in the data frame from start to end (non-inclusive)
        
        #Extract wavelength and intensity values from each row
        wavelength_value = df['Wavelength'][row] 
        intensity_value = df['Intensity'][row]
        
        #Perform calcuations and add to variables that will later allow us to calculate CoM
        int_wvln += wavelength_value * intensity_value
        total_int += intensity_value
        
        #print('Wvln: ', wavelength_value,'Int: ', intensity_value)
        
    #Print statements to check that code is working correctly     
    print('Int x Wvln: ', int_wvln, 'Total intensity: ', total_int)
    
    #Calculate center of mass
    center_of_mass = int_wvln / total_int
    print(center_of_mass)
    print('End of frame', frame_number)
    
    #Add the data point (time, CoM) for the group to the list of x,y values to plot
    time_stamps.append(frame_number)
    com_values.append(center_of_mass)
    #print(time_stamps)
    #print(com_values)
    
    #Update variables and then return to the top of the loop
    frame_number +=1 
    start = (frame_number - 1) * pixel_size
    end = start + pixel_size

print('Time values: ', time_stamps)
print('COM values: ', com_values)
    
#Plot the data    
plt.plot(time_stamps, com_values, marker='o', linestyle='-',)
plt.xlabel('Time (s)')
plt.ylabel('Center of Mass (nm)')
plt.title('Example File')


```

### Without wavlength restriction
This code is for plotting the CoM for the large file. It calculates the center of mass using all data points from the frame (using wavelengths and intensities from wavelength 400 - 980 nm). As I saw later when plotting PL spectra, the lack of adding a wavelength restriction to calculate CoM can be problematic if the laser signal is significantly high in intensity, as it can blueshift the CoM. This is why I wrote another code below that has wavelength restriction. 

```python
#Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/Karen/Desktop/GitHub/PL_Tools/11-20-2023/11-20-2023 14_33_34 488nm_5mW_od1p5_CsPbI2Br_sample8_1000ms.csv'

#Read csv file into a dataframe
df = pd.read_csv(file_path)

#Define file length and pixel (group) size
file_length = len(df)
pixel_size = 1340 
print('File length is',file_length,'rows.')


#Initiate variables 
frame_number = 1
start = 0 #start at row index 0 
end = 1340 #end at row index 1340 (not included) 

#Create lists to hold data points (x and y values)
time_stamps = []
com_values = []

#Create conditional loop to go through each frame and calculate CoM
while start < file_length:
    
    #Initiate variables for CoM calcuation 
    int_wvln = 0
    total_int = 0
    
    for row in range(start,end): #Loop through each row in the frame
        
        #Extract wavelength and intensity values for CoM calculation
        wavelength_value = df['Wavelength'][row]
        intensity_value = df['Intensity'][row]
        
        #Update variables for CoM calculation 
        int_wvln += wavelength_value * intensity_value
        total_int += intensity_value
    
    center_of_mass = int_wvln / total_int
    #print('End of frame', frame_number)
    
    
        
    #Add the data point (time, com) for the group to the list of x,y values
    time_stamps.append(frame_number)
    com_values.append(center_of_mass)
    #print(time_stamps)
    #print(com_values)
    
    #Update variables 
    frame_number +=1 
    start = (frame_number - 1) * pixel_size
    end = start + pixel_size

#Plot data    
plt.plot(time_stamps, com_values)
plt.xlabel('Time (s)')
plt.ylabel('Center of Mass (nm)')
plt.title('Center of Mass vs Time for I2Br Film (No Wavelength Restriction)')

```

### With wavelength restriction (550-750nm)

This code is for plotting the CoM for the large file. It calculates the center of mass using data points only between 550 - 750 nm from each frame. This restriction is helpful for dealing with intense PL signal due to the laser, which can skew the CoM value (making CoM appear smaller than it really is). 

```python
#Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt

#Replace with your file path here
file_path = '/Users/Karen/Desktop/GitHub/PL_Tools/11-20-2023/11-20-2023 14_33_34 488nm_5mW_od1p5_CsPbI2Br_sample8_1000ms.csv'

#Read csv file into data frame
df = pd.read_csv(file_path)

#Define file length and pixel (group) size
file_length = len(df)
pixel_size = 1340 
print('File length is',file_length,'rows.')


#Initiate variables 
frame_number = 1
start = 0 #start at row index 0
end = 1340 #end at row index 1340 (not included) 

#Create lists that will hold x,y values for plotting on the graph
time_stamps = []
com_values = []

#Create conditional loop to go through each frame and calculate CoM
while start < file_length:
    
    #Initiate variables for CoM calculation 
    int_wvln = 0
    total_int = 0
    
    for row in range(start,end): #Loop through each row in the frame 
        
        #Extract wavelength and intensity values from the row
        wavelength_value = df['Wavelength'][row]
        intensity_value = df['Intensity'][row]
        
        #Conditional statement for using wavelengths between 550nm and 750nm to calculate CoM
        if wavelength_value >= 550 and wavelength_value <= 750:
            #print(wavelength_value)
            
            #Update variables for CoM calculation
            int_wvln += wavelength_value * intensity_value
            total_int += intensity_value
    center_of_mass = int_wvln / total_int
    #print('End of frame', frame_number)
    
    
        
    #Add the data point (time, com) for the group to the list of x,y values
    time_stamps.append(frame_number)
    com_values.append(center_of_mass)
    #print(time_stamps)
    #print(com_values)
    
    #Update variables 
    frame_number +=1 
    start = (frame_number - 1) * pixel_size
    end = start + pixel_size

#Plot data points
plt.plot(time_stamps, com_values)
plt.xlabel('Time (s)')
plt.ylabel('Center of Mass (nm)')
plt.title('Center of Mass vs Time for I2Br Film (With Wavelength Restriction)')


```

This is code below is for if you want to plot the data but zoomed into a specific time stamp.  

```python
#Imagine a simpler example
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/Karen/Desktop/GitHub/PL_Tools/11-20-2023/11-20-2023 14_33_34 488nm_5mW_od1p5_CsPbI2Br_sample8_1000ms.csv'

df = pd.read_csv(file_path)

#Define file length and pixel (group) size
file_length = len(df)
pixel_size = 1340 
print('File length is',file_length,'rows.')


#Initiate variables 
frame_number = 1
start = 0 #start at row index 1 
end = 1340 #end at row index 3 (not included) 

time_stamps = []
com_values = []

#Create conditional loop to go through each frame and calculate CoM
while start < file_length:
    int_wvln = 0
    total_int = 0
    for row in range(start,end): 
        wavelength_value = df['Wavelength'][row]
        intensity_value = df['Intensity'][row]
        int_wvln += wavelength_value * intensity_value
        total_int += intensity_value
    center_of_mass = int_wvln / total_int
    #print('End of frame', frame_number)
    
    
        
    #Add the data point (time, com) for the group to the list of x,y values only for time 60 - 300s
    if frame_number > 60: #Plotting only points from 1 minute (60s) onward 
        time_stamps.append(frame_number)
        com_values.append(center_of_mass)
    #print(time_stamps)
    #print(com_values)
    
    #Update variables 
    frame_number +=1 
    start = (frame_number - 1) * pixel_size
    end = start + pixel_size
    
plt.plot(time_stamps, com_values)
plt.xlabel('Time (s)')
plt.ylabel('Center of Mass (nm)')
plt.title('Center of Mass vs Time for I2Br Film (With Time Restriction)')

```
